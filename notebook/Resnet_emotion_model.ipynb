```python
from keras.applications.resnet50 import ResNet50
from keras.layers import Input,Activation
from keras.layers import Conv2D,Dense,MaxPool2D
from keras.layers import Flatten,BatchNormalization
from keras.layers import Dropout
import keras
```

```python
new_input = Input(shape=(56,56,3))
resnet = ResNet50(input_tensor=new_input)
```

```python
resnet.summary()
```

```python
model = keras.models.Model(inputs = resnet.input,outputs = [resnet.get_layer('avg_pool').output])
```

```python
flatten = Flatten()(model.layers[-1].output)
dense_1 = Dense(1024,activation=keras.activations.relu)(flatten)
drop_1 = Dropout(0.1)(dense_1)
dense_2 = Dense(512,activation=keras.activations.relu)(drop_1)
norm_2 = BatchNormalization()(dense_2)
dense_3 = Dense(256,activation=keras.activations.relu,kernel_regularizer= keras.regularizers.l2(1e-3))(norm_2)
norm_3 = BatchNormalization()(dense_3)
dense_4 = Dense(128,activation=keras.activations.relu)(norm_3)
norm_4 = BatchNormalization()(dense_4)
drop_4 = Dropout(0.1)(norm_4)
dense_5 = Dense(64,activation= keras.activations.relu)(drop_4)
norm_5 = BatchNormalization()(dense_5)
drop_5 = Dropout(0.1)(norm_5)
dense_6 = Dense(32,activation=keras.activations.relu)(drop_5)
emotion = Dense(7,activation='softmax',name='emotion')(dense_6)
```

```python
new_model = keras.models.Model(inputs = model.input,outputs=emotion)
```

```python
new_model.summary()
```

```python
for layer in new_model.layers:
    if layer.name == 'avg_pool':
        break
    else:
        layer.trainable =False
```
